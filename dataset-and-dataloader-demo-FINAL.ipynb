```python
from sklearn.datasets import make_classification
import torch
```

```python
# Step 1: Create a synthetic classification dataset using sklearn
X, y = make_classification(
    n_samples=10,       # Number of samples
    n_features=2,       # Number of features
    n_informative=2,    # Number of informative features
    n_redundant=0,      # Number of redundant features
    n_classes=2,        # Number of classes
    random_state=42     # For reproducibility
)
```

```python
# Convert the data to PyTorch tensors
X = torch.tensor(X, dtype=torch.float32)     # convert numpy arrays into Tensors
y = torch.tensor(y, dtype=torch.long)
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataset(Dataset):

  def __init__(self, features, labels):
    """ Initialize the dataset with features and labels. """
    self.features = features
    self.labels = labels

  def __len__(self):
    """ Return the number of rows in the dataset. """
    return self.features.shape[0]

  def __getitem__(self, index):
    """ Return the features and labels for a given index.
        Also you can do any kind of transformation in this function before return.
        For example: images database: crop, resize, brightness etc are transformation of datas
    """

    return self.features[index], self.labels[index]
```

```python
dataset = CustomDataset(X, y)
```

```python
len(dataset)
```

```python
dataset[0]
```

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=False)     # we dont want to shuffle the rows. so shuffle=False
```

```python
for batch_features, batch_labels in dataloader:

  print(batch_features)
  print(batch_labels)
  print("-"*50)
```
