```python
# create model class
import torch
import torch.nn as nn

class Model(nn.Module):
  """This is very basix example of creating model class"""
  def __init__(self, num_features):

    super().__init__()                              # calling __init__() method of parent class nn.Module
    self.linear = nn.Linear(num_features, 1)        # z = wx + b (used linear layer)
    self.sigmoid = nn.Sigmoid()                     # sigmoid activation function

  def forward(self, features):

    out = self.linear(features)       # here getting z = wx + b
    out = self.sigmoid(out)           # here getting sigmoid(z) .... dz/dy

    return out
```

```python
# create model class
import torch
import torch.nn as nn

class Model(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    """
    THIS RED CODES ARE WORKING TOGETHER IN FORWARD METHOD AND MODEL CLASS. SO UNCOMMENT BOTH TOGETHER

    self.linear1 = nn.Linear(num_features, 3)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()
    """
    # Let's sy if we have more than 1000 inputs(features) and we need to create neural networks. In this case, we have container concept.
    # we will use "Sequential" containers to make it customized and short

    self.network = nn.Sequential(
        nn.Linear(num_features, 3),
        nn.ReLU(),
        nn.Linear(3, 1),
        nn.Sigmoid()
    )


  def forward(self, features):
    """
    out = self.linear1(features)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)
    """
    # If using Sequential container, just pass features into self.network variable.
    out = self.network(features)


    return out
```

```python
# create dataset
features = torch.rand(10,5)         # 5 columns taken because we took 5 features in neural network

features
```

```python
# create model
model = Model(features.shape[1])
```

```python
# call model for forward pass
# model.forward(features)       # we called this way in Python but it is advisable to use as below. just call the function by attributes
model(features)
```
